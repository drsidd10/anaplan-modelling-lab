## 1. Create a List
Equivalent to an Anaplan List containing members such as Products, Regions, Cost Centers, etc.

```python
products = List("Products", ["Core", "Premium", "EV"])
products
```

## 2. Create a Subset
In Anaplan, subsets are used to reduce dimensionality and improve performance (Planual guidance).

```python
subset = ListSubset(products, [1, 0, 1])
subset.items
```

## 3. Define a Hierarchy
Parent → Child relationships, just like in Anaplan composite hierarchies.

```python
parents = {
    "Core": "Vehicles",
    "Premium": "Vehicles",
    "EV": "Vehicles"
}

hierarchy = Hierarchy("Product Hierarchy", parents, None)
hierarchy.parents_of("Core"), hierarchy.children_of("Vehicles")
```

## 4. Time Testing (using Time Engine)
PREVIOUS, CUMULATE, and TIMESUM mirror Anaplan time formulas.

```python
from src.time.time import PREVIOUS, CUMULATE, TIMESUM, make_time_range

time = make_time_range(periods=4)
series = pd.Series([100, 150, 200, 300], index=time)

previous = PREVIOUS(series)
cumulative = CUMULATE(series)
total = TIMESUM(series, time[0], time[-1])

previous, cumulative, total
```
