# 09 — Visualisations for Connected Planning
This notebook generates example charts for Revenue, EBITDA, FCF, Workforce Cost, and Inventory.

```python
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
import numpy as np
```

## Sample Synthetic Data (Replace with Actual Consolidated Outputs)

```python
scenarios=['Base','Upside','Downside']
revenue=[120,135,110]
ebitda=[30,38,25]
fcf=[18,25,12]
workforce_cost=[15,16,14]
inventory=[3000,3400,2800]
```

## Revenue by Scenario

```python
plt.figure(figsize=(6,4))
plt.bar(scenarios,revenue,color=['grey','green','red'])
plt.title('Revenue by Scenario')
plt.xlabel('Scenario')
plt.ylabel('Revenue (M)')
plt.tight_layout()
plt.savefig('revenue_chart.png')
'Generated revenue_chart.png'
```

## EBITDA Comparison

```python
plt.figure(figsize=(6,4))
plt.plot(scenarios,ebitda,marker='o')
plt.title('EBITDA by Scenario')
plt.xlabel('Scenario')
plt.ylabel('EBITDA (M)')
plt.tight_layout()
plt.savefig('ebitda_chart.png')
'Generated ebitda_chart.png'
```

## FCF Bridge Style (Simple)

```python
plt.figure(figsize=(6,4))
plt.bar(scenarios,fcf,color=['blue','green','orange'])
plt.title('Free Cash Flow by Scenario')
plt.xlabel('Scenario')
plt.ylabel('FCF (M)')
plt.tight_layout()
plt.savefig('fcf_chart.png')
'Generated fcf_chart.png'
```

## Workforce Cost

```python
plt.figure(figsize=(6,4))
plt.bar(scenarios,workforce_cost,color=['purple','cyan','grey'])
plt.title('Workforce Cost by Scenario')
plt.xlabel('Scenario')
plt.ylabel('Cost (M)')
plt.tight_layout()
plt.savefig('workforce_chart.png')
'Generated workforce_chart.png'
```

## Inventory Ending Balance

```python
plt.figure(figsize=(6,4))
plt.plot(scenarios,inventory,marker='s')
plt.title('Ending Inventory by Scenario')
plt.xlabel('Scenario')
plt.ylabel('Inventory Units')
plt.tight_layout()
plt.savefig('inventory_chart.png')
'Generated inventory_chart.png'
```
